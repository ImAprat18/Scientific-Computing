```python
plt.grid()
ax=plt.gca()
req_rectangle=plt.Rectangle((0, 0), 100, 100, fill=False, color='green', linewidth=2)
req_circle=plt.Circle((50, 50), 50, fill=False, color='red', linewidth=2)
ax.add_patch(req_rectangle)
ax.add_patch(req_circle)
ax.set_aspect('equal')
plt.xlim([-10, 110])
plt.ylim([-10, 110])
plt.show()
```

```python
in_ctr=0
out_ctr=0
for i in range(1000):
    x=random.randint(0,100)
    y=random.randint(0,100)
    plt.scatter(x,y)
    dist_sq=(x-50)**2 + (y-50)**2
    dist=math.sqrt(dist_sq)
    if dist<=50:
        in_ctr+=1
    else:
        out_ctr+=1
print(in_ctr, out_ctr)
print(4*(in_ctr/(in_ctr+out_ctr)))
area_c=math.pi*50*50
area_s=100*100
print(4*(area_c/area_s))
ax=plt.gca()
req_rectangle=plt.Rectangle((0, 0), 100, 100, fill=False, color='green', linewidth=3)
req_circle=plt.Circle((50, 50), 50, fill=False, color='white', linewidth=3)
ax.add_patch(req_rectangle)
ax.add_patch(req_circle)
ax.set_aspect('equal')
plt.xlim([-10, 110])
plt.ylim([-10, 110])
plt.show()
```
